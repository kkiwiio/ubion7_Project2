```python
import pandas as pd
```

```python
main_df = pd.read_excel('./main_df_dropna_completed.xlsx',index_col=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 샘플 데이터 생성
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# 상위 1%와 하위 1%의 값을 구합니다.
upper_cutoff = np.percentile(main_df['R&D집중도(%)'], 99)
lower_cutoff = np.percentile(main_df['R&D집중도(%)'], 1)
```

```python
print(upper_cutoff)
print(lower_cutoff)
```

```python
# 윈저라이징 적용
data_winsorized = np.clip(main_df['R&D집중도(%)'], lower_cutoff, upper_cutoff)

# boxplot 비교를 위한 subplot 생성
plt.figure(figsize=(10, 5))

# 원본 데이터의 boxplot
plt.subplot(1, 2, 1)
plt.boxplot(main_df['R&D집중도(%)'])
plt.title('원본 데이터의 Boxplot')
plt.ylabel('값')

# 윈저라이징 후 데이터의 boxplot
plt.subplot(1, 2, 2)
plt.boxplot(data_winsorized)
plt.title('윈저라이징 후 데이터의 Boxplot')
plt.ylabel('값')

plt.tight_layout()
plt.show()
```

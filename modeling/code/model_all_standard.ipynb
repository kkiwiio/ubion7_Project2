```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
################################################
##         LASSO, SMOTE, Logistic            ##
###############################################
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import ADASYN
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.inspection import permutation_importance

models = [
        DecisionTreeClassifier(),
        LogisticRegression(),
        RandomForestClassifier(),
        XGBClassifier(),
        LGBMClassifier()]

params = [[0.01,0.03],
        ['auto',0.33]]

rdict_smote={'models':[],'params':[],'accuracy':[],'precision':[],'recall':[],'f1_score':[], 'cv_score':[]}
rdict_ada={'models':[],'params':[],'accuracy':[],'precision':[],'recall':[],'f1_score':[], 'cv_score':[]}

for clf in models:
        for C in params[0]:
                for strategy in params[1]:
                        ## Feature Selection
                        lasso = SelectFromModel(LogisticRegression(C=C,penalty='l1',solver='saga'))
                        lasso.fit(x_train_scaled, y_train)

                        lasso_support = lasso.get_support()
                        feature_selected = x_train_scaled.loc[:,lasso_support].columns.tolist()
                        # feature_selected.append('산업부도율')

                        x_test_feature_selected = x_test[feature_selected]

                        #####################################################################################

                        ## OverSampling_1
                        sm = SMOTE(random_state=0, sampling_strategy=strategy)
                        x_train_smote, y_train_smote = sm.fit_resample(x_train_scaled[feature_selected], y_train)

                        clf_1 = clf.fit(x_train_smote, y_train_smote)
                        pred_1 = clf.predict(x_test_feature_selected)

                        cv_scores_1 = cross_val_score(clf_1, x_train_smote, y_train_smote, cv=5, scoring='recall')

                        # 성능 지표 계산
                        accuracy_1 = accuracy_score(y_test, pred_1)
                        precision_1 = precision_score(y_test, pred_1)
                        recall_1 = recall_score(y_test, pred_1)
                        f1_1 = f1_score(y_test, pred_1)

                        rdict_smote['models'].append((clf)); 
                        rdict_smote['params'].append((C,strategy)); 
                        rdict_smote['accuracy'].append(round(accuracy_1,2))
                        rdict_smote['precision'].append(round(precision_1,2))
                        rdict_smote['recall'].append(round(recall_1,2))
                        rdict_smote['f1_score'].append(round(f1_1,2))
                        rdict_smote['cv_score'].append(round(cv_scores_1.mean(), 2))

                        # Permutation Importance 계산
                        print(f'Model : {clf_1} & C : {C} & sampling_strategy : {strategy} --- SMOTE \n')
                        perm_importance = permutation_importance(clf_1, x_test_feature_selected, y_test, n_repeats=30, random_state=0)

                        # 각 피처의 Permutation Importance 값 출력
                        for i, feature_name in enumerate(x_test_feature_selected.columns):
                                print(f"{feature_name}: {perm_importance.importances_mean[i]}")

                        

                        print("____________________________________________________________________")
                        ######################################################################################

                        ## OverSampling_2
                        adasyn = ADASYN(random_state=0, sampling_strategy=strategy)
                        x_train_ad, y_train_ad = adasyn.fit_resample(x_train_scaled[feature_selected], y_train)

                        clf_2 = clf.fit(x_train_ad, y_train_ad)
                        pred_2 = clf.predict(x_test_feature_selected)

                        cv_scores_2 = cross_val_score(clf_2, x_train_ad, y_train_ad, cv=5, scoring='recall')

                        # 성능 지표 계산
                        accuracy_2 = accuracy_score(y_test, pred_2)
                        precision_2 = precision_score(y_test, pred_2)
                        recall_2 = recall_score(y_test, pred_2)
                        f1_2 = f1_score(y_test, pred_2)
                        
                        rdict_ada['models'].append((clf_2)); 
                        rdict_ada['params'].append((C,strategy)); 
                        rdict_ada['accuracy'].append(round(accuracy_2,2))
                        rdict_ada['precision'].append(round(precision_2,2))
                        rdict_ada['recall'].append(round(recall_2,2))
                        rdict_ada['f1_score'].append(round(f1_2,2))
                        rdict_ada['cv_score'].append(round(cv_scores_2.mean(), 2))
                        
                        print(f'Model : {clf_2} & C : {C} & sampling_strategy : {strategy} --- Ada \n')
                        # Permutation Importance 계산
                        perm_importance = permutation_importance(clf_2, x_test_feature_selected, y_test, n_repeats=30, random_state=0)

                        # 각 피처의 Permutation Importance 값 출력
                        for i, feature_name in enumerate(x_test_feature_selected.columns):
                                print(f"{feature_name}: {perm_importance.importances_mean[i]}")


smote_result = pd.DataFrame(data=rdict_smote)
ada_result = pd.DataFrame(data=rdict_ada)
```

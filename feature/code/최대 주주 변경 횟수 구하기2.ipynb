```python
#기업의 년도별 최대 주주 이름을 확인하기 위해 groupby 후 성명을 list로 변경
new_a = largest_shareholder.groupby(["회사명", "거래소코드", "회계년도"])["성명"].apply(set)
```

```python
df_b = pd.DataFrame(new_a)
```

```python
df = df_b.reset_index()
df
```

```python
 #"다음년도_성명_변경여부" 컬럼을 생성하여 이전 컬럼과 한명이라도 최대 주주가 바뀌었는지 여부 확인
df['다음년도_성명_변경여부'] = (df['성명'] != df['성명'].shift(1))
```

```python
# 마지막 년도는 다음 년도가 없으므로 비교하지 않기 위해 해당 컬럼을 False (0)로 설정합니다.
df.loc[df.groupby(['회사명', '거래소코드']).head(1).index, '다음년도_성명_변경여부'] = False
```

```python
#원본데이터에서 회사명, 거래소코드, 회계년도,회계년도만 다시 가져오기
largest_shareholder_2 = largest_shareholder[['회사명', '거래소코드', '회계년도', '회계년월']]
#똑같이 중복값 제거
largest_shareholder_2 = largest_shareholder_2.drop_duplicates()
```

```python
#largest_shareholder_2["성명"]컬럼을 새로 만든 후 df_a의 성명리스트 값 저장
largest_shareholder_2["성명"] = df_a
```

```python
 #"다음년도_성명_변경여부" 컬럼을 생성하여 이전 컬럼과 한명이라도 최대 주주가 바뀌었는지 여부 확인
largest_shareholder_2['다음년도_성명_변경여부'] = (largest_shareholder_2['성명'] != largest_shareholder_2['성명'].shift(1))
```

```python
# 마지막 년도는 다음 년도가 없으므로 비교하지 않기 위해 해당 컬럼을 False (0)로 설정합니다.
largest_shareholder_2.loc[largest_shareholder_2.groupby(['회사명', '거래소코드']).head(1).index, '다음년도_성명_변경여부'] = False
```

```python
largest_shareholder_2
```

```python
# "회사명"과 "거래소코드"를 기준으로 그룹화하고, "다음년도_성명_변경여부"의 누적합을 구하여 "최대주주변경 횟수"를 계산
largest_shareholder_2['최대주주변경 횟수'] = largest_shareholder_2.groupby(['회사명', '거래소코드'])['다음년도_성명_변경여부'].cumsum()
```

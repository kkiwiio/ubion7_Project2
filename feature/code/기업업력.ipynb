```python
import pandas as pd
```

```python
df = pd.read_excel('./메인 데이터 프레임2.xlsx')
df = df.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1)
```

```python
from datetime import datetime
```

```python
# 회계년도와 상장일 컬럼의 데이터를 시계열 형식으로 변환
df['회계년도'] = pd.to_datetime(df['회계년도'], format='%Y/%m')
df['상장일'] = pd.to_datetime(df['상장일'], format='%Y/%m/%d')
```

```python
# 회계년도에서 상장일을 뺀 값을 계산하여 새로운 컬럼인 '업력'을 생성
df['업력'] = (df['회계년도'] - df['상장일']).dt.days
```

```python
# '업력' 컬럼에 저장된 값은 일단위로 되어있음-> 연도 단위로 변경
df['업력'] = df['업력'] // 365
```

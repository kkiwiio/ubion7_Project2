```python
df = pd.read_excel('./기업제외 최대주주.xlsx')
```

```python
df
```

```python
df =df.drop(['Unnamed: 0'],axis=1)
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
a = df.groupby(["회사명", "거래소코드", "회계년도"])["성명"].apply(set)
```

```python
df_a = pd.DataFrame(a)
```

```python
df_a  
```

```python
 #"다음년도_성명_변경여부" 컬럼을 생성하여 이전 컬럼과 한명이라도 최대 주주가 바뀌었는지 여부 확인
df_a['다음년도_성명_변경여부'] = (df_a['성명'] != df_a['성명'].shift(1))
```

```python
df_a
```

```python
df_a.loc[df_a.groupby(['회사명', '거래소코드']).head(1).index, '다음년도_성명_변경여부'] = False
```

```python
# "회사명"과 "거래소코드"를 기준으로 그룹화하고, "다음년도_성명_변경여부"의 누적합을 구하여 "최대주주변경 횟수"를 계산
df_a['최대주주변경 횟수'] = df_a.groupby(['회사명', '거래소코드'])['다음년도_성명_변경여부'].cumsum()
```

```python
df_a = df_a.reset_index()
```

```python
df_a.to_excel('./최대주주 변경 최종.xlsx')
```

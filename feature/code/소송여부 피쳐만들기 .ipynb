```python
import pandas as pd
```

```python
lawsuit = pd.read_excel('../data/소송.xlsx')
```

```python
lawsuit = lawsuit.drop_duplicates()
```

```python
lawsuit = lawsuit.rename(columns={'구분(10-피고 20-원고 90-기타)(IFRS)':'소송'})
```

```python
lawsuit['소송여부'] = lawsuit['소송'].fillna(1).apply(lambda x: 0 if x in [10, 20, 90] else 1)
```

```python
lawsuit = lawsuit[['회사명','거래소코드','회계년도','소송여부']]
```

```python
lawsuit['회계년도'] = pd.to_datetime(lawsuit['회계년도'],format='%Y/%m')
```

```python
main_df = pd.read_excel('../code/비재무 메인 데이터.xlsx',index_col=0)
main_df
```

```python
main_df = pd.merge(left=main_df,right=lawsuit,on=['회사명','거래소코드','회계년도'],how='left')
```

```python
main_df.to_excel('../비재무 피쳐 데이터(완).xlsx')
```

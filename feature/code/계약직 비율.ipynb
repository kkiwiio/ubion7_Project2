```python
import pandas as pd
```

```python
person = pd.read_excel('../data/계약직 비율.xlsx')
```

```python
person
```

```python
person['기간제비율'] = person.groupby(['회사명', '회계년도'])['기간제_기말인원'].transform(lambda x: x / person['합계_기말인원'] * 100)
```

```python
person['기간제비율'] = person['기간제비율'].round(2)
```

```python
person['회계년도'] =pd.to_datetime(person['회계년도'],format='%Y/%m')
```

```python
person = person[['회사명','거래소코드','회계년도','기간제비율']]
```

```python
main_df = pd.read_excel('./메인데이터프레임2.xlsx',index_col=0)
main_df
```

```python
main_df = pd.merge(left=main_df,right=person,on=['회사명','거래소코드','회계년도'],how='left')
```

```python
main_df.to_excel('./비재무 메인 데이터.xlsx')
```

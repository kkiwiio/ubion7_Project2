```python
import pandas as pd
```

```python
df = pd.read_excel('../data/윤아_비재무.xlsx')
df.head(3)
```

```python
rd_df= df[['지출액_연결','지출액_개별']]
```

```python
rd_df
```

```python
rd_df['지출액_연결'] = rd_df['지출액_연결'].fillna(rd_df['지출액_개별'])
```

```python
rd_df
```

```python
rd_df = rd_df['지출액_연결']
```

```python
revenu_df = df[['매출액_연결','매출액_연결.1']]
```

```python
revenu_df
```

```python
revenu_df['매출액_연결'] = revenu_df['매출액_연결'].fillna(revenu_df['매출액_연결.1'])
```

```python
revenu_df = revenu_df['매출액_연결']
```

```python
result = pd.concat([rd_df,revenu_df],axis=1)
result
```

```python
result  = result.fillna(0)
```

```python
result['R&D비용'] = (result['지출액_연결'] / result['매출액_연결']) * 100
```

```python
result['R&D비용'] = result['R&D비용'] .round(2)
```

```python
result.to_excel('r&d 개발 최종.xlsx')
```

```python
export_df= df[['수출액_연결','수출액_개별']]
export_df
```

```python
export_df['수출액_연결'] = export_df['수출액_연결'].fillna(0)
```

```python
export_df = export_df['수출액_연결']
```

```python
result2 = pd.concat([export_df,revenu_df],axis=1)
result2
```

```python
result2['수출기업여부'] = (result2['수출액_연결'] / result2['매출액_연결'] * 100 > 50).astype(int)
```

```python
result2.to_excel('./수출기업여부 최종.xlsx')
```

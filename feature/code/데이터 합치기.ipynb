```python
import pandas as pd
```

```python
main_df = pd.read_excel('../data/메인 데이터 프레임2.xlsx',index_col=0)
main_df.head()
```

```python
main_df.info()
```

```python
x =pd.read_excel('./임원진 주식 보유 최종.xlsx',index_col=0)
x.head(5)
```

```python
x.info()
```

```python
x['회계년도'] = pd.to_datetime(x['회계년도'], format='%Y/%m')
```

```python
x = x[['거래소코드','회계년도','임원진 주식 보유']]
```

```python
main_df = pd.merge(left=main_df,right=x,how='left',on = ['거래소코드','회계년도'])
```

```python
main_df
```

```python
y = pd.read_excel('./최대주주 변경 최종.xlsx',index_col=0)
y.head(5)
```

```python
y['회계년도'] = pd.to_datetime(y['회계년도'], format='%Y/%m')
y = y[['거래소코드','회계년도','최대주주변경 횟수']]
```

```python
main_df = pd.merge(left=main_df,right=y,on=['거래소코드','회계년도'],how='left')
```

```python
import pandas as pd
```

```python
person_df = pd.read_excel('../data/직원증감.xlsx')
```

```python
person_df
```

```python
person_df['이전_년도_기말인원'] = person_df.groupby(["회사명", "거래소코드"])["합계_기말인원"].shift(1)
```

```python
person_df['직원수 증감'] = ((person_df['합계_기말인원'] - person_df['이전_년도_기말인원']) / person_df['이전_년도_기말인원'])*100
```

```python
person_df['직원수 증감'] = person_df['직원수 증감'].round(2)
```

```python
person_df = person_df[['회사명','거래소코드','회계년도','직원수 증감']]
```

```python
person_df['회계년도'] = pd.to_datetime(person_df['회계년도'],format='%Y/%m')
```

```python
main_df = pd.read_excel('./메인데이터프레임.xlsx',index_col=0)
main_df 
```

```python
main_df = pd.merge(left=main_df,right=person_df,on=['회사명','거래소코드','회계년도'],how='left')
```

```python
main_df.to_excel('./메인데이터프레임2.xlsx')
```

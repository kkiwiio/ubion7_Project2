```python
import pandas as pd
```

```python
change_df = pd.read_excel('../data/회계감사인현황.xlsx')
```

```python
change_df
```

```python
change_df = change_df.drop_duplicates()
change_df.info()
```

```python
a =change_df.groupby(["회사명", "거래소코드", "회계년도"])["회계감사법인"].apply(list)
```

```python
a
```

```python
df_a = pd.DataFrame(a)
df_a
```

```python
df_a['다음년도_감사법인_변경여부'] = (df_a['회계감사법인'] != df_a['회계감사법인'].shift(1))
```

```python
df_a
```

```python
df_a.loc[df_a.groupby(['회사명', '거래소코드']).head(1).index, '다음년도_감사법인_변경여부'] = False
```

```python
df_a['감사법인 변경 횟수'] = df_a.groupby(['회사명', '거래소코드'])['다음년도_감사법인_변경여부'].cumsum()
```

```python
df_a = df_a.reset_index()
```

```python
df_a['회계년도'] = pd.to_datetime(df_a['회계년도'], format='%Y/%m')
```

```python
df_a = df_a[['회사명','거래소코드','회계년도','감사법인 변경 횟수']]
```

```python
main_df = pd.read_excel('./메인데이터프레임.xlsx',index_col=0)
main_df
```

```python
main_df = pd.merge(left=main_df,right=df_a,on=['회사명','거래소코드','회계년도'],how='left')
```

```python
main_df.to_excel('./메인데이터프레임.xlsx')
```
